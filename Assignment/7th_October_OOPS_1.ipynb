#### Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

```python
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance, password):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance
        self.password = password
    def validate(self):
        if self.account_number == "123456" and self.password == "abc":
            return True
        else:
            return False
    def deposit(self, amount_deposited):
        if self.validate():
            print("Current balance: ", self.balance)

            self.balance += amount_deposited
            print("New balance: ", self.balance)
        else:
            print("Invalid credentials")
    def withdraw(self, amount_withdrawn):
        if self.validate() and self.balance!=0 and amount_withdrawn<= self.balance:
            print("Current balance: ", self.balance)
            self.balance -= amount_withdrawn
            print("New balance: ", self.balance)          
        else:
            print("Invalid credentials or insufficient balance or zero balance")
account1 = BankAccount("123456", "pwskills", 1000, "abc")

amount_deposited = int(input("Enter amount to deposit: "))
account1.deposit(amount_deposited)
print("\n")
amount_withdrawn = int(input("Enter amount to be withdraw: "))
account1.withdraw(amount_withdrawn)
```

#### Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

```python
class Book:
    def __init__(self, title, borrowed):
        self.title = title
        self.borrowed = borrowed
class Library:
    def __init__(self):
        self.books = [Book("Life of PI",False), 
                                Book("Lord of the Rings",False),
                                Book("Harry Potter",False),
                                Book("Divergent", False),
                                Book("Insurgent", False)]
    def display_all_books(self):
        print("Available Books:\n")
        for book in self.books:
            print(book.title, end = " ")
            print(":",book.borrowed)
       
    def add_books(self):
        n = int(input("Enter number of books to add: "))
        for i in range(n):
            book = input("Enter book to add: ")
            book_obj = Book(book, False)
            self.books.append(book_obj)
            print(book, " has been added")
        print("\n")
        self.display_all_books()
        
    def borrow_book(self):
        n = int(input("Enter number of book to be borrowed: "))
        if n<=len(self.books):
            for i in range(n):
                book = input("Enter title of book to be borrowed: ")
                is_available = True
                for item in self.books:
                    if item.title == book and item.borrowed == False:
                        item.borrowed = True
                        is_available = False
                        print(item.title, " has been borrowed")
                if is_available:
                    print("Book not available")           
        else:
            print("Number of books entered exceeds number of books available")
            return
        self.display_all_books()

lib = Library()
lib.display_all_books()
print("\n")
lib.add_books()
print("\n")
lib.borrow_book()
```

#### Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.

```python
class Product:
    def __init__(self, title, quantity):
        self.title = title
        self.quantity = quantity

class Inventory:
    def __init__(self):
        self.products = [Product("Laptop",1), Product("Dektop",2), Product("Camera",1), Product("GPU",4)]
    
    def display_all_products(self):
        
        for product in self.products:
            print(product.title, ": ", product.quantity)
    def add_products(self):
        n = int(input("Enter number of products to be added: "))
        
        for i in range(n):
            
            product_title = input("Enter title of product the product to be added: ")
            is_available = False
            product_quantity = int(input("Enter quantity of the product to be added:"))
            for item in self.products:
                if item.title == product_title:
                    is_available = True
                    
                    item.quantity += product_quantity
            
            if not is_available:
                
                self.products.append(Product(product_title,product_quantity))
        
        self.display_all_products()
    def update_product_quantity(self):
        product_title = input("Enter title of product whose quantity needs to be updated: ")
        for item in self.products:
            if item.title == product_title: 
                choice = input("Enter 1 if quantity needs to be increased and 0 if quantity needs to be decreased: ")
                if choice == '1':
                    add = int(input("Enter quantity to be added: "))
                    item.quantity += add
                elif choice == '0':
                    subtract = int(input("Enter quantity to be subtracted: "))
                    if subtract<=item.quantity: 
                        item.quantity -= subtract
                    else:
                        print("Product is of Insufficient quantity")
        self.display_all_products()
        
inventory = Inventory()
inventory.display_all_products()   
inventory.add_products()
inventory.update_product_quantity()
```

#### Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

```python
class Cuboid:
    def __init__(self, length, width, height):
        self.length = length
        self.width = width
        self.height = height
    def calculate_Area(self):
        area = 2*(self.length*self.width + self.length*self.height + self.width*self.height)
        print("Area of cuboid: ", area)
    def calculate_Perimeter(self):
        perimeter = 4*(length + width + height)
        print("Perimeter of cuboid: ", perimeter)
length = int(input("Enter length of cuboid: "))
width = int(input("Enter width of cuboid: "))
height = int(input("Enter height of cuboid: "))
        
cuboid = Cuboid(length, width, height)
cuboid.calculate_Area()
cuboid.calculate_Perimeter()
```

#### Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

```python
class Post:
    def __init__(self, title, description):
        self.title = title
        self.description = description
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = [Post("PWskills", "Python course"), Post("ineuron", "java course"), Post("PWskills", "ML course")]
    def display_posts(self):
        print("\n")
        for post in self.posts:
            
            print("Post title: ", post.title)
            print("Post description: ", post.description)
        print("\n")
    def add_posts(self):
        n = int(input("Enter number of posts that you want to add: "))
        for i in range(n):
            title = input("Enter post title: ")
            description = input("Enter post description: ")
            self.posts.append(Post(title,description))
        print("\n")
        print("Updated List of posts: ")  
        self.display_posts()
   
            
    def search_posts_by_keyword(self):
        exists = False
        print("\n")
        keyword = input("Enter keyword: ")
        for post in self.posts:
            if keyword in post.title:
                print("Post(s) found")
                print("Post title: ", post.title)
                print("Post description: ", post.description)
                exists = True
        if not exists:
            print("Post does not exist!")

username = input("Enter username: ")
profile = SocialMediaProfile(username)
profile.display_posts()
profile.add_posts()  

profile.search_posts_by_keyword()    
print("\n")
profile.search_posts_by_keyword()  
```

#### Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
import datetime
class Task:
    def __init__(self, title, description, due_date):
        self.title = title
        self.description = description
        self.due_date = due_date
        self.isCompleted = False
        
class ToDoList:
    def __init__(self):
        self.tasks = [Task("Python HW", "Write Python code", datetime.datetime(2023, 11,1)),
                      Task("ML HW", "Write mathematical intuition for Linear Regression", datetime.datetime(2023,11,5)),
                      Task("SQL HW", "Write SQL queries for the given problem statement", datetime.datetime(2023,11,10))    
                     ]
    def display_all_tasks(self):
        
        print("List of all tasks: ")
        for task in self.tasks:
                print("Task title: ", task.title)
                print("Task description: ", task.description)
                print("Task Due date: ", task.due_date)
                if task.isCompleted:
                    
                    print("Status: Completed")
                else:
                    print("Status: Incomplete")
        print("\n")
            
    def display_pending_tasks(self):
        print("Pending Tasks: \n")
        for task in self.tasks:
            if task.isCompleted == False:
                print("Task title: ", task.title)
                print("Task description: ", task.description)
                print("Task Due date: ", task.due_date)
                print("Status: Incomplete")
        
            print("\n")
    
    def add_tasks(self):
        n = int(input("Enter number of tasks that you want to add: "))
        for i in range(n):
            title = input("Enter task title: ")
            description = input("Enter task description: ")
            year = int(input("Enter task due date year: "))
            month = int(input("Enter task due date month: "))
            day = int(input("Enter task due date day: "))
            self.tasks.append(Task(title, description, datetime.datetime(year, month,day)))
        print("\n")
        print("Updated list of tasks: ")
        self.display_all_tasks()
    def mark_tasks_as_completed(self):
        n = int(input("Enter number of tasks that you want to mark as completed:"))
        for i in range(n):
            title = input("Enter task title: ")
            for task in self.tasks:
                if task.title == title:
                    task.isCompleted = True
        print("\n")
        print("Updated list of tasks: ")
        self.display_all_tasks()
        print("\n")
            
todo = ToDoList()
todo.display_all_tasks()
todo.add_tasks()
todo.mark_tasks_as_completed()
```
